```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
import seaborn as sns
import scipy.stats as stast
```

```python
grp = ['Data Scientist', 'ML Engineer', 'DL Engineer', 'FullStack Developer', 'Statiscian']

val = [24, 18, 14, 32, 20]
```

```python
plt.style.use('ggplot')

font_path = '/content/CalSans-Regular.ttf'

font_prop = fm.FontProperties(fname=font_path)
fm.fontManager.addfont(font_path)

plt.rcParams['font.family'] = font_prop.get_name()

plt.figure(figsize=(16, 12))

colores = sns.color_palette('coolwarm')

plt.get_cmap('Spectral')
```

```python
fig, ax = plt.subplots()
plt.title('Profesiones Tech')
ax.bar(grp, val, color=colores, edgecolor='black', linewidth=3.5, alpha=1, label=grp)
plt.legend()
plt.tight_layout()
plt.xlabel('Profesiones', fontweight='bold', color='black')
plt.ylabel('Salario', fontweight='bold', color='black')
plt.show()
```

```python
fig, ax = plt.subplots()
plt.title('Profesiones en Tech')
ax.pie(val, labels=grp, autopct='%1.1f%%', colors=colores, explode=(0.1, 0, 0, 0, 0), shadow=True)
plt.legend(labels=grp)
plt.tight_layout()
plt.show()
```

```python
meses = ["Junio", "Julio", "Marzo", "Diciembre", "Febrero"]
ventas = [100, 200, 180, 400, 300]

plt.plot(meses, ventas)
plt.xlabel("Meses", fontweight="bold", color="black")
plt.ylabel("Paquetes vendidos", fontweight="bold", color="black")
plt.title("Grfico de ventas")
plt.legend(["Ventas mensuales.", "Nmero de ventas"])
plt.show()
```

```python
# Semilla para la reproducibilidad
np.random.seed(5)

# Simulacin de datos
x = np.random.normal(0, 2, 100)

# Grfico de violn
fig, ax = plt.subplots()
parts = ax.violinplot(x, showmeans=False, showmedians=False, showextrema=False)
for pc in parts['bodies']:
    pc.set_facecolor('#D43F3A')
    pc.set_edgecolor('black')
    pc.set_alpha(1)
ax.set_title("Grfico de violn")
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
plt.show()
```
